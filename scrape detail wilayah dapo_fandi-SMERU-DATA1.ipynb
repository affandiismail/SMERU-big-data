```python
data_wilayah = []

for prov in data_prov:
    print(f'Mengambil data provinsi {prov["nama"]}')
    data_kab_request = requests.get(f'https://dapo.kemdikbud.go.id/rekap/dataSekolah?id_level_wilayah={prov["id_level_wilayah"]}&kode_wilayah={prov["kode_wilayah"]}&semester_id=20222')
    data_kab = data_kab_request.json()
    
    for kab in data_kab:
        print(f'Mengambil data kabupaten {kab["nama"]}')
        data_kec_request = requests.get(f'https://dapo.kemdikbud.go.id/rekap/dataSekolah?id_level_wilayah={kab["id_level_wilayah"]}&kode_wilayah={kab["kode_wilayah"]}&semester_id=20222')
        data_kec = data_kec_request.json()
        
        for kec in data_kec:
            print(f'Mengambil data kecamatan {kec["nama"]}')
            data_wilayah.append({
                    'prov': prov['nama'],
                    'kode_prov' : prov['kode_wilayah'],
                    'kab' : kab['nama'],
                    'kode_kab' : kab['kode_wilayah'],
                    'kec' : kec['nama'],
                    'kode_kec' : kec['kode_wilayah'],
            })
```

```python
df_wilayah = pd.DataFrame('data_wilayah', dtype = 'str')
```
